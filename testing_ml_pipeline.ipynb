```python
import pandas as pd
import random
import numpy as np
import string
from sklearn.datasets import make_classification

# Generate dummy classification dataset
X, y = make_classification(
    n_samples=1500,
    n_features=10,
    n_informative=5,
    n_redundant=2,
    n_classes=2,
    random_state=42
)


# Create DataFrame with feature names
feature_names = [f'feature_{i}' for i in range(X.shape[1])]
df = pd.DataFrame(X, columns=feature_names)

# Add target column
df['target'] = y

# Function to generate a random 5-character ID
def generate_id():
    return ''.join(random.choices(string.ascii_letters, k=5))

# Create the "id" column
df['id_col'] = [generate_id() for _ in range(len(df))]

# Move 'id' column to the front
df = df[['id_col'] + ['target'] + feature_names]

df['feature_10'] = [random.choice(['a', 'b', 'c', np.nan]) for _ in range(len(df))]

df_train = df.iloc[:1000].copy()
df_test = df.iloc[1000:].copy()

# Save to CSV
df_train.to_csv('example_inputs_v1.csv', index=False)
df_test.to_csv('example_inputs_test_v1.csv', index=False)

display(df_train.head())
display(df_test.head())
```
